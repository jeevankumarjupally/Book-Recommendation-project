## Item - Item Similarity

```python
from scipy.sparse import csr_matrix
rating_sparse = csr_matrix(rating_matrix.values)
item_similarity = cosine_similarity(rating_sparse.T,dense_output=False)
```

```python
def recommend_books(book_name, n=10, min_similarity=0.2):

    matches = [col for col in rating_matrix.columns if col.lower() == book_name.lower()]
    if not matches:
        return f"'{book_name}' not found in dataset"

    book_col = matches[0]

    item_index = rating_matrix.columns.get_loc(book_col)
    sim_scores = item_similarity[item_index].toarray().ravel()

    top_id = np.argsort(sim_scores)[::-1]
    top_id = [i for i in top_id if i != item_index]

    filtered_id = [i for i in top_id if sim_scores[i] > min_similarity][:n]
    top_books = rating_matrix.columns[filtered_id]
    top_scores = sim_scores[filtered_id]

    rec_books = pd.DataFrame({
        "Book-Title": top_books,
        "Similarity": top_scores
    })

    if "Book-Author" in books.columns:
        rec_books = pd.merge(
            rec_books,
            books[['Book-Title', 'Book-Author']].drop_duplicates(),
            on="Book-Title",
            how="left"
        )

    return rec_books.drop_duplicates().head(n)
b = input("Enter Book Name: ")
res = recommend_books(b, n=2)

print('_____________________________________________________________________________')
print("                            THE RECOMMENDED BOOKS                            ")
print('_____________________________________________________________________________')
res
```

# hybrid recommendation (user and item together)

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix
# ======================
# STEP 1: Prepare Data
# ======================
# Assume you have two DataFrames: ratings and books
# ratings: User-ID, ISBN, Book-Rating
# books: ISBN, Book-Title, Book-Author, Year-Of-Publication, Publisher, Location, Age

# Create user-item matrix
rating_matrix = data.pivot_table(index="User-ID",
                                    columns="Book-Title",
                                    values="Book-Rating")

rating_matrix = rating_matrix.fillna(0)

# User similarity
user_similarity = cosine_similarity(rating_matrix)
user_similarity_df = pd.DataFrame(user_similarity,
                                  index=rating_matrix.index,
                                  columns=rating_matrix.index)

# Item similarity
sparse_matrix = csr_matrix(rating_matrix.values)
item_similarity = cosine_similarity(sparse_matrix.T, dense_output=False)
item_similarity_df = pd.DataFrame(item_similarity.toarray(),
                                  index=rating_matrix.columns,
                                  columns=rating_matrix.columns)

# ======================
# STEP 2: User-Based Recommendation Function
# ======================
def recommend_items_user_based(user_id, n=5):
    if user_id not in rating_matrix.index:
        return pd.DataFrame({"Book-Title": [], "Score_UserCF": []})

    # Find similar users
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:11].index
    similar_users_ratings = rating_matrix.loc[similar_users].mean().sort_values(ascending=False)

    # Exclude already rated items
    items_already_rated = rating_matrix.loc[user_id]
    recommended = similar_users_ratings[items_already_rated == 0].head(n)

    return pd.DataFrame(recommended).reset_index().rename(columns={0: "Score_UserCF"})

# ======================
# STEP 3: Item-Based Recommendation Function
# ======================
def recommend_items_item_based(user_id, n=5):
    if user_id not in rating_matrix.index:
        return pd.DataFrame({"Book-Title": [], "Score_ItemCF": []})

    user_ratings = rating_matrix.loc[user_id]
    rated_books = user_ratings[user_ratings > 0].index

    scores = pd.Series(dtype=float)

    for book in rated_books:
        similar_books = item_similarity_df[book].sort_values(ascending=False).drop(book)
        scores = scores.add(similar_books, fill_value=0)

    # Exclude already rated
    scores = scores.drop(rated_books, errors="ignore")

    return pd.DataFrame(scores.sort_values(ascending=False).head(n)).reset_index().rename(columns={0: "Score_ItemCF"})
# ======================
# STEP 4: Hybrid Recommendation
# ======================
def recommend_hybrid(user_id, n=5, alpha=0.6):
    """
    alpha = weight for user-based CF (0 <= alpha <= 1)
    (1-alpha) = weight for item-based CF
    """
    user_recs = recommend_items_user_based(user_id, n*2)   # get more to blend
    item_recs = recommend_items_item_based(user_id, n*2)

    # Merge both
    hybrid = pd.merge(user_recs, item_recs, on="Book-Title", how="outer").fillna(0)

    # Final hybrid score
    hybrid["Hybrid_Score"] = alpha * hybrid["Score_UserCF"] + (1-alpha) * hybrid["Score_ItemCF"]

    # Sort and merge with book details
    hybrid = hybrid.sort_values("Hybrid_Score", ascending=False).head(n)
    hybrid = pd.merge(hybrid, books[["Book-Title","Book-Author","Publisher"]],
                      on="Book-Title", how="left").drop_duplicates()

    return hybrid[["Book-Title","Hybrid_Score","Book-Author","Publisher"]]

# ======================
# STEP 5: Example
# ======================
user_id = int(input("Enter User ID: "))
print("\nHybrid Recommendations:\n")
print(recommend_hybrid(user_id, n=5, alpha=0.6))
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(data['book_content'])
```

```python
def recommend_books(book_title, top_n=1):
    # Get index of the given book
    idx = data[data['Book-Title'] == book_title].index[0]

    # Compute similarity scores for this book vs all others
    cosine_sim = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

    # Enumerate all similarity scores
    sim_scores = list(enumerate(cosine_sim))

    # Sort by similarity (excluding the book itself)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get top-N similar books (skip the first one = the same book)
    sim_scores = sim_scores[1:top_n+1]

    # Extract book indices
    book_indices = [i[0] for i in sim_scores]

    return data.iloc[book_indices]['Book-Title'].values
```

### Evaluation Metrics

For evaluating the recommendation system, we will consider the following metrics:

- **Mean Squared Error (MSE):** This metric is suitable for evaluating the performance of a recommendation system when the task is to predict the rating a user would give to a book (regression problem). A lower MSE indicates better prediction accuracy.

- **Precision, Recall, and Accuracy:** These metrics are useful when the recommendation task is framed as a classification problem, where we predict whether a user will like a book or not (e.g., recommending books with a rating above a certain threshold).
    - **Precision:** Measures the proportion of recommended books that are actually relevant to the user. High precision means fewer irrelevant recommendations.
    - **Recall:** Measures the proportion of relevant books that are successfully recommended to the user. High recall means fewer missed relevant recommendations.
    - **Accuracy:** Measures the overall correctness of the recommendations (proportion of correct predictions).

The choice of which metrics to prioritize will depend on the specific goals of the recommendation system (e.g., whether it's more important to avoid irrelevant recommendations or to ensure all relevant books are recommended).

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

print("Training data shape:", train_data.shape)
print("Testing data shape:", test_data.shape)
```

### Evaluate User-Based Collaborative Filtering

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# Create train and test rating matrices
train_rating_matrix = train_data.pivot_table(index="User-ID", columns="Book-Title", values="Book-Rating").fillna(0)
test_rating_matrix = test_data.pivot_table(index="User-ID", columns="Book-Title", values="Book-Rating").fillna(0)

# Align test matrix with train matrix columns
test_rating_matrix = test_rating_matrix.reindex(columns=train_rating_matrix.columns, fill_value=0)

# Calculate user similarity on the training data
user_similarity_train = cosine_similarity(train_rating_matrix)
user_similarity_df_train = pd.DataFrame(user_similarity_train,
                                        index=train_rating_matrix.index,
                                        columns=train_rating_matrix.index)

# Function to predict rating for a user and book using user-based CF
def predict_user_based(user_id, book_title, train_matrix, user_sim_df):
    if user_id not in train_matrix.index or book_title not in train_matrix.columns:
        return 0  # Cannot predict

    similar_users = user_sim_df[user_id].sort_values(ascending=False)[1:11].index
    similar_users = [u for u in similar_users if u in train_matrix.index] # Ensure similar user is in train matrix

    if not similar_users:
        return 0 # No similar users found in training data

    # Calculate weighted average of ratings from similar users
    weighted_sum = 0
    similarity_sum = 0
    for sim_user in similar_users:
        rating = train_matrix.loc[sim_user, book_title]
        similarity = user_sim_df.loc[user_id, sim_user]
        weighted_sum += similarity * rating
        similarity_sum += similarity

    return weighted_sum / similarity_sum if similarity_sum > 0 else 0

# Evaluate User-Based CF on the test set
actual_ratings = []
predicted_ratings_user_based = []

# Iterate through the test data to get predictions
# This might be slow for large datasets, consider optimizing if needed
for index, row in test_data.iterrows():
    user_id = row['User-ID']
    book_title = row['Book-Title']
    actual_rating = row['Book-Rating']

    predicted_rating = predict_user_based(user_id, book_title, train_rating_matrix, user_similarity_df_train)

    actual_ratings.append(actual_rating)
    predicted_ratings_user_based.append(predicted_rating)

# Calculate RMSE for User-Based CF
rmse_user_based = sqrt(mean_squared_error(actual_ratings, predicted_ratings_user_based))
print(f"User-Based CF RMSE: {rmse_user_based}")
```

### Evaluate Item-Based Collaborative Filtering

```python
# Calculate item similarity on the training data
item_similarity_train = cosine_similarity(train_rating_matrix.T)
item_similarity_df_train = pd.DataFrame(item_similarity_train,
                                        index=train_rating_matrix.columns,
                                        columns=train_rating_matrix.columns)

# Function to predict rating for a user and book using item-based CF
def predict_item_based(user_id, book_title, train_matrix, item_sim_df):
    if user_id not in train_matrix.index or book_title not in train_matrix.columns:
        return 0  # Cannot predict

    user_ratings = train_matrix.loc[user_id]
    rated_books = user_ratings[user_ratings > 0].index

    if book_title not in item_sim_df.index:
        return 0 # Cannot predict if book is not in item similarity matrix

    # Calculate weighted average of ratings for similar items
    weighted_sum = 0
    similarity_sum = 0
    for rated_book in rated_books:
        if rated_book in item_sim_df.index: # Ensure rated book is in item similarity matrix
            similarity = item_sim_df.loc[book_title, rated_book]
            rating = user_ratings.loc[rated_book]
            weighted_sum += similarity * rating
            similarity_sum += similarity

    return weighted_sum / similarity_sum if similarity_sum > 0 else 0

# Evaluate Item-Based CF on the test set
actual_ratings = []
predicted_ratings_item_based = []

# Iterate through the test data to get predictions
# This might be slow for large datasets, consider optimizing if needed
for index, row in test_data.iterrows():
    user_id = row['User-ID']
    book_title = row['Book-Title']
    actual_rating = row['Book-Rating']

    predicted_rating = predict_item_based(user_id, book_title, train_rating_matrix, item_similarity_df_train)

    actual_ratings.append(actual_rating)
    predicted_ratings_item_based.append(predicted_rating)

# Calculate RMSE for Item-Based CF
rmse_item_based = sqrt(mean_squared_error(actual_ratings, predicted_ratings_item_based))
print(f"Item-Based CF RMSE: {rmse_item_based}")
```

### Evaluate Hybrid Recommendation

```python
# Function to predict rating using the hybrid model
def predict_hybrid(user_id, book_title, train_matrix, user_sim_df, item_sim_df, alpha=0.6):
    user_based_pred = predict_user_based(user_id, book_title, train_matrix, user_sim_df)
    item_based_pred = predict_item_based(user_id, book_title, train_matrix, item_sim_df)

    # If both predictions are 0, return 0
    if user_based_pred == 0 and item_based_pred == 0:
        return 0

    # Weighted average prediction
    return alpha * user_based_pred + (1 - alpha) * item_based_pred

# Evaluate Hybrid CF on the test set
actual_ratings = []
predicted_ratings_hybrid = []

# Iterate through the test data to get predictions
# This might be slow for large datasets, consider optimizing if needed
for index, row in test_data.iterrows():
    user_id = row['User-ID']
    book_title = row['Book-Title']
    actual_rating = row['Book-Rating']

    predicted_rating = predict_hybrid(user_id, book_title, train_rating_matrix, user_similarity_df_train, item_similarity_df_train)

    actual_ratings.append(actual_rating)
    predicted_ratings_hybrid.append(predicted_rating)

# Calculate RMSE for Hybrid CF
rmse_hybrid = sqrt(mean_squared_error(actual_ratings, predicted_ratings_hybrid))
print(f"Hybrid CF RMSE: {rmse_hybrid}")
```

**Important Considerations for Deployment:**

*   **Model Complexity and Interpretability:** Hybrid models can be more complex to understand and maintain compared to simpler collaborative filtering methods.
*   **Real-time vs. Batch Processing:** Consider whether recommendations need to be generated in real-time or can be pre-calculated in batches. This will influence your deployment strategy.

In conclusion, while the Hybrid model shows promising accuracy, a thorough evaluation considering various metrics and performance on a larger scale is recommended before final deployment.
